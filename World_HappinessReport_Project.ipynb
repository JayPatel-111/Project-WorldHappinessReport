```python
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mode
import seaborn as sns
```

```python
# Importing and reading CSV and using .head() to see first 5 rows
happiness_data = pd.read_csv('Downloads/world_happiness_report_2022.csv')
happiness_data.head()
```

```python
happiness_data.isna().sum()
```

```python
happiness_data.dtypes
```

# Statistical Analysis using mean, median, count and mode

```python
# below is the mean happiness score for all data in the column Happiness score
mean_happiness_score = happiness_data['Happiness score'].mean()
mean_happiness_score
```

```python
# below is the median happiness score for all data in the column Happiness score
median_happiness_score = happiness_data['Happiness score'].median()
median_happiness_score
```

```python
# count for the score
count_happiness_score = happiness_data['Happiness score'].count()
count_happiness_score
```

```python
# mode 
mode_happiness_score = mode(happiness_data['Happiness score'])
mode_happiness_score
```

## Patterns/relationships 

```python
corr_gdp_happiness = happiness_data['Happiness score'].corr(happiness_data['Explained by: GDP per capita'])
corr_gdp_happiness
```

```python
corr_social_support_happiness = happiness_data['Happiness score'].corr(happiness_data['Explained by: Social support'])
corr_social_support_happiness
```

```python
corr_freedom_happiness = happiness_data['Explained by: Freedom to make life choices'].corr(happiness_data['Happiness score'])
corr_freedom_happiness
```

## Visualization below

```python
# below is two graphs showing relations between happiness score and GDP per cap and social support.
plt.scatter(happiness_data['Happiness score'], happiness_data['Explained by: GDP per capita'])
plt.xlabel('Happiness Score')
plt.ylabel('GDP per capita')
plt.title(' Happiness Score VS GDP per capita')
plt.show()

plt.scatter(happiness_data['Happiness score'], happiness_data['Explained by: Social support'])
plt.xlabel('Happiness Score')
plt.ylabel('Social Support')
plt.title('Happiness Score VS Social Support')
plt.show()
```

## Statistical Visualization for bar graph, violin graph and histogram showing same data but different graphs

```python
# Statisticsal visualization 
plt.boxplot(happiness_data['Happiness score'])
plt.xlabel('Happiness Score')
plt.title('Distribution of Happiness Score')
plt.show()
```

```python
sns.violinplot(data=happiness_data, y='Happiness score')
plt.ylabel('Happiness Score')
plt.title('Distribution of Happiness Score')
plt.show()
```

```python
happiness_scores = happiness_data['Happiness score']
plt.hist(happiness_scores, bins=10, edgecolor='blue')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Score')
plt.show()
```
